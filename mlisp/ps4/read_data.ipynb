```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
with open("clean_signal.txt") as f:
    reader = csv.reader(f)
    #next(reader) # skip header
    #data = [float(r) for r in reader]
    #data = [r[0].split() for r in reader]
    data = []
    for r in reader:
        row = r[0].split()
        rownum = [float(el) for el in row]
        data.append(rownum)
    #print(next(reader)[0].split())
data = np.array(data)
```

```python
with open("noisy_signal.txt") as f:
    reader = csv.reader(f)
    #next(reader) # skip header
    #data = [float(r) for r in reader]
    #data = [r[0].split() for r in reader]
    datan = []
    for r in reader:
        row = r[0].split()
        rownum = [float(el) for el in row]
        datan.append(rownum)
    #print(next(reader)[0].split())
datan = np.array(datan)
```

```python
plt.plot(data[:,3])
plt.show()
```
